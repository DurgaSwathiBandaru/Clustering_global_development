```python
Data=Df.copy()
Data.head()
```

```python
Data=Data.drop(columns=['Number of Records'])
```

```python
Data['Business Tax Rate'] = Data['Business Tax Rate'].astype(str).str.replace('[%,]', '', regex=True)
Data['Business Tax Rate'] = pd.to_numeric(Data['Business Tax Rate'], errors='coerce')

Data['GDP'] = Data['GDP'].astype(str).str.replace('[$,]', '', regex=True)
Data['GDP'] = pd.to_numeric(Data['GDP'], errors='coerce')

Data['Health Exp/Capita'] = Data['Health Exp/Capita'].astype(str).str.replace('[$,]', '', regex=True)
Data['Health Exp/Capita'] = pd.to_numeric(Data['Health Exp/Capita'], errors='coerce')

Data['Tourism Outbound'] = Data['Tourism Outbound'].astype(str).str.replace('[$,]', '', regex=True)
Data['Tourism Outbound'] = pd.to_numeric(Data['Tourism Outbound'], errors='coerce')

Data['Tourism Inbound'] = Data['Tourism Inbound'].astype(str).str.replace('[$,]', '', regex=True)
Data['Tourism Inbound'] = pd.to_numeric(Data['Tourism Inbound'], errors='coerce')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Data['Country_enc']=le.fit_transform(Data['Country'])
Data['Country_enc']=Data['Country_enc'].astype(float)
Data.drop(['Country'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Scaled_Data=scaler.fit_transform(Data)
```

```python
Scaled_Data
```

```python
from sklearn.decomposition import PCA
pca=PCA()
pca_components=pca.fit_transform(Scaled_Data)
explained_variance=pca.explained_variance_
cumulative_explained_variance = np.cumsum(explained_variance)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Cumulative Explained Variance by PCA")
plt.axhline(y=0.90, color='r', linestyle='--', label="90% Variance Threshold")
plt.legend()
plt.show()
```

```python
n_components=2
pca=PCA(n_components)
pca
```

```python
pca_data = pca.fit_transform(Scaled_Data)
pca_data = pd.DataFrame(pca_data, columns=[f'PC{i+1}' for i in range(pca_data.shape[1])])
pca_data.head()
```

```python
from sklearn.cluster import  KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
wcss = []
k_values = range(2, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k,init='k-means++', random_state=0)
    kmeans.fit(pca_data)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.plot(range(2,11), wcss, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5,random_state=42) 
kmeans_labels=kmeans.fit_predict(pca_data)
```

```python
kmeans_silhouette = silhouette_score(pca_data, kmeans_labels)
kmeans_db_index=davies_bouldin_score(pca_data,kmeans_labels)
print(f"Silhouette Score for K-Means (K={5}): {kmeans_silhouette:.3f}")
print(f"David bouldin Score for K-Means (K={5}): {kmeans_db_index:.3f}")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_data.iloc[:, 0], y=pca_data.iloc[:, 1], hue=kmeans_labels, palette='viridis')
plt.title("Kmeans Clustering")
plt.legend(title="Cluster")
plt.show()
```

```python
# 2. Hierarchical Clustering
linked = linkage(pca_data, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10,6))
dendrogram(linked, truncate_mode='lastp', p=25, leaf_rotation=0, leaf_font_size=8)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data points")
plt.ylabel("Distance")
plt.show()
```

```python
hierarchical=AgglomerativeClustering(n_clusters=3)
hierarchical_labels=hierarchical.fit_predict(pca_data)
```

```python
hierarchical_silhouette = silhouette_score(pca_data, hierarchical_labels)
hierarchical_db_index=davies_bouldin_score(pca_data,hierarchical_labels)
print(f"Silhouette Score for Hierarchical clustering (K={3}): {hierarchical_silhouette:.3f}")
print(f"David bouldin Score for K-Means (K={3}): {hierarchical_db_index:.3f}")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_data.iloc[:, 0], y=pca_data.iloc[:, 1], hue=hierarchical_labels, palette='viridis')
plt.title("Hierarchical Clustering")
plt.legend(title="Cluster")
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
k=5
neighbor = NearestNeighbors(n_neighbors=k).fit(pca_data)
distance,index = neighbor.kneighbors(pca_data)
```

```python
k_distances=np.sort(distance[:,k-1])
plt.figure(figsize=(10,6))
plt.plot(k_distances,marker="o",linestyle="-",label="k-distance")
plt.axhline(y=0.8,color="green",linestyle="-")
plt.xlabel("Data points sorted by Distance")
plt.ylabel(f"K-Distance Value")
plt.title(f"Distance Plot for Dbscan")
plt.grid(True)
plt.show()
```

```python
dbscan= DBSCAN(eps=0.8, min_samples=4)
dbscan_labels= dbscan.fit_predict(pca_data)
```

```python
dbscan_silhouette = silhouette_score(pca_data, dbscan_labels)
dbscan_db_index = davies_bouldin_score(pca_data,dbscan_labels )
print(f"Silhouette Score for Dbscan clustering : {dbscan_silhouette:.3f}")
print(f"David bouldin Score for Dbscan: {dbscan_db_index:.3f}")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_data.iloc[:, 0], y=pca_data.iloc[:, 1], hue=dbscan_labels, palette='viridis')
plt.title("Dbscan Clustering")
plt.legend(title="Cluster")
plt.show()
```

```python
print("\nComparison and Analysis:")
print(f"Silhouette Score :- Kmeans: {kmeans_silhouette}, Hierarchical: {hierarchical_silhouette}, Dbscan:{dbscan_silhouette}")
print(f"Davies-Bouldin Index :-Kmeans: {kmeans_db_index}, Hierarchical: {hierarchical_db_index}, Dbscan:{dbscan_db_index}")
```

```python
fig,axes=plt.subplots(1,3,figsize=(20,8))

def plt_clusters(ax, labels, title):
    scatter = ax.scatter(pca_data.iloc[:,0],pca_data.iloc[:,1], c=labels, cmap='viridis', edgecolor='w', alpha=0.7)
    ax.set_title(title,fontsize=16)
    ax.set_xlabel("Principal Component 1",fontsize=14)
    ax.set_ylabel("Principal Component 2",fontsize=14)
    ax.legend(*scatter.legend_elements(),title="Clusters",fontsize=12)
                       
plt_clusters(axes[0], kmeans_labels,"K-Means Clustering on PCA Data") 
plt_clusters(axes[1], hierarchical_labels,"Hierarchical Clustering on PCA Data")  
plt_clusters(axes[2], dbscan_labels,"DBSCAN Clustering on PCA Data")   

plt.tight_layout()
plt.show()
```
